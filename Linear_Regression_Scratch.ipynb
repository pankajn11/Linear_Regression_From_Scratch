```python
import pandas as pd
import numpy as np
import random
import ssl
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python

ssl._create_default_https_context = ssl._create_unverified_context

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"

columns = [ "mpg","cylinders", "displacement", "horsepower","weight","acceleration","model_year","origin","car_name"]

df = pd.read_csv( url, names=columns, na_values="?",sep=r"\s+" )

df.head()
```

```python
df.fillna(df['horsepower'].mean(),inplace = True )
df.isna().sum()
```

```python
features = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin']
X = df[features]
Y = df['mpg']
```

```python
#X
```

### Train Test Split

```python

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state = 42)
```

### Standard-Scaler

```python
scale_ob  = StandardScaler()
X_train = scale_ob.fit_transform(X_train)
X_test = scale_ob.transform(X_test)

#---- Transforming Y_train as well thought standard scaler-------------
Y_train = scale_ob.fit_transform(Y_train.values.reshape(-1,1)).ravel()
Y_test = scale_ob.transform(Y_test.values.reshape(-1,1)).ravel()



print( 'X_train --> {}     Y_train --> {}\nX_test  --> {}       Y_test --> {} '.format(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape))
```

### Linear Regression Sk-Learn

```python
LR_obj = LinearRegression()
LR_obj.fit(X_train,Y_train)
```

```python
#Y_pred = np.round(LR_obj.predict(X_train),1)
Y_pred = LR_obj.predict(X_train)
```

```python
#Y_train = np.round(Y_train)
```

```python
# Intercept and Weight 

print('Intercept  = {} '.format(LR_obj.intercept_))
print()
print('Weights rounded off to two digits = {} '.format(np.round(LR_obj.coef_,2)))
```

```python
# MSE 
print(f'mean_squared_error {round(mean_squared_error(Y_train,Y_pred),4)}')
```
