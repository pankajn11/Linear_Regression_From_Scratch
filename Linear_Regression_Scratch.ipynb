#### Importing libraries

```python
import pandas as pd
import numpy as np
import random
import ssl
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

#### Reading 'auto-mpg' dataset using pandas

```python

ssl._create_default_https_context = ssl._create_unverified_context

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"

columns = [ "mpg","cylinders", "displacement", "horsepower","weight","acceleration","model_year","origin","car_name"]

df = pd.read_csv( url, names=columns, na_values="?",sep=r"\s+" )

#df.head()
```

```python
df.fillna(df['horsepower'].mean(),inplace = True )
df.isna().sum()
```

```python
Y = df['mpg']
X = df.drop(columns=['mpg','car_name'])
```

```python
del df
```

```python
#X.head()
```

#### Train Test Split

```python

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state = 42)
```

#### standardizing the features using Standard-Scaler

```python
scale_ob  = StandardScaler()
X_train = scale_ob.fit_transform(X_train)
X_test = scale_ob.transform(X_test)

#---- Transforming Y_train as well thought standard scaler-------------
Y_train = scale_ob.fit_transform(Y_train.values.reshape(-1,1)).ravel()
Y_test = scale_ob.transform(Y_test.values.reshape(-1,1)).ravel()

print( 'X_train --> {}     Y_train --> {}\nX_test  --> {}       Y_test --> {} '.format(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape))
```

#### Linear Regression using SkLearn

```python
LR_obj = LinearRegression()
LR_obj.fit(X_train,Y_train)
```

```python
Y_pred = LR_obj.predict(X_train)
```

```python
# Intercept and Weight 
SKL_weights = LR_obj.coef_
SKL_intercept  = LR_obj.intercept_
print('Intercept  = {} '.format(LR_obj.intercept_))
print()
print('Weights rounded off to two digits = {} '.format(np.round(SKL_weights,2)))
```

```python
# MSE 

SKL_MSE = mean_squared_error(Y_train,Y_pred)

print(f'mean_squared_error {round(SKL_MSE,4)}')
```

### Linear Regression Our Implementation

```python
epoch = 5000
X = X_train
Y_a = Y_train 
b=0 
W=np.zeros(X_train.shape[1])
lr  = 0.1

min_cost_dict = {}

for e in range(epoch):

    Y_p = np.dot(X,W)+b
    error = Y_a-Y_p
    MSE =  np.mean(error**2)

    if e % 100 == 0:
        print('Epoch -> {} ,MSE -> {}  ,W -> {} ,b -> {} '.format(e,MSE,np.round(W,2),b))
        print()

    # capturing the Weights and Intercept for each MSE , so we can get the Weights and Intercept for the minimum value of MSE
    min_cost_dict[MSE] = (epoch,W,b)
    
    # Updating Weight and Intercept
    dW = (-2 / X.shape[0]) * np.dot(X.T, error)
    W =  W - lr*dW

    db = (-2/X.shape[0])*np.sum(error)
    b = b - lr*db

    

    
```

```python
OUR_MSE = min(min_cost_dict.keys())
OUR_intercept = min_cost_dict[min(min_cost_dict.keys())][2]
OUR_weights = min_cost_dict[min(min_cost_dict.keys())][1]
```

```python

# Intercept and Weight from our implemenatation VS Sklearn

print(f'"Intercept" Sklearn = {SKL_intercept} ,\n\n"Intercept" our implementation = {OUR_intercept}')
print()
print(f'"Weights" Sklearn = \n{SKL_weights} ,\n\n"Weights" our implementation =\n {OUR_weights}')
print()
print(f'"MSE" Sklearn = {SKL_MSE} ,\n\n"MSE" our implementation = {OUR_MSE}')
```
